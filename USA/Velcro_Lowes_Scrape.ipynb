```python
from bs4 import BeautifulSoup
from selenium import webdriver

browser = webdriver.Chrome()
url = 'https://www.lowes.com/search?searchTerm=velcro'
browser.get(url)
innerHTML = browser.execute_script('return document.body.innerHTML')
soup = BeautifulSoup(innerHTML)

data = []
```

```python
items = soup.find_all("li", class_='product-wrapper')

for item in range(len(items)):
    data += [{'product-image' : items[item].find(class_='product-image').find('img')['src'],
              'product-name' : items[item].find(class_='product-title').get_text(strip=True)
             }]
    data[-1]['product-name'] = data[-1]['product-name'].replace('\t','')
```
