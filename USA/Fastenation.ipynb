```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time

# Going to Velcro search page in Home Depot
options = webdriver.ChromeOptions()
#options.add_argument("--headless")
browser = webdriver.Chrome(executable_path=r"chromedriver.exe",chrome_options=options)
#url = 'https://www.fastenation.com/'
url = 'https://www.fastenation.com/catalogsearch/result/?q=velcro'
browser.get(url)
```

```python
data = []

hasNext = True
y = 0
multiplier = .025

pheight = browser.execute_script("return document.body.scrollHeight") * multiplier
scrollto = 0

#print("made all variables")

while hasNext:
    try:        
        lastY = scrollto
        #print("lastY modified")
        scrollto = scrollto + pheight 
        #print("Height: " + str(scrollto))
        browser.execute_script("window.scrollTo(0, " + str(scrollto) + ");")       
        #print('loaded more!')
        time.sleep(0.1)
        #loadResults += 1
        if(scrollto >= browser.execute_script("return document.body.scrollHeight")):
            print("Found bottom of page!")
            hasNext = False
    except:
        print("error!")
        hasNext = False

# Get all HTML in document and assign it to BeautifulSoup variable
innerHTML = browser.execute_script('return document.body.innerHTML')
soup = BeautifulSoup(innerHTML,"lxml")

# Finds all of the individual products
items = soup.find_all('a', 'product-image eQ')
```

```python
# Get all product information on this page
for index in range(len(items)):
    item = items[index]
    item_img = item.find('img')
    #item_info = item.find('div','product-info')
    src = item_img.get('src', item_img.get('dfr-src'))
    if src is None:
        continue
    else:
        #item_name = (item.find('div','product-info').find('a',{'target' : '_parent'}).text)
        #item_name = (item.find('div','product-info').find('a').text)
        item_name = item['title']        
        data += [{'product-image': src,
                  'product-name' : item_name.replace('\n', '').replace('\t','').strip()}]
```

```python
for i in range(len(data)):
    print(i+1)
    print(data[i]['product-image'])
    print(data[i]['product-name'])
    
browser.quit()
```
