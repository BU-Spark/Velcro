```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time

# Going to Velcro search page in Home Depot
options = webdriver.ChromeOptions()
#options.add_argument("--headless")
browser = webdriver.Chrome(executable_path=r"chromedriver.exe",chrome_options=options)
#url = 'https://www.staples.com/'
url = 'https://www.staples.com/velcro/directory_velcro?deptFid=Department_3A_22Velcro_26reg_3B_22'
browser.get(url)
```

```python
data = []

hasNext = True

loadResults = 0
maxLoadResults = int(input('What is the maximum number of times you would like the program to load more results: '))

while hasNext:
    try:
        wait = WebDriverWait(browser, 10)
        loadMoreBtn = wait.until(EC.presence_of_element_located((By.ID, 'load-more-results')))        
        actions = ActionChains(browser)
        actions.move_to_element(loadMoreBtn).perform()
        loadMoreBtn.click()
        #print('Clicked!')
        time.sleep(1)
        loadResults += 1
        if(loadResults >= maxLoadResults):
            hasNext = False
    except:
        hasNext = False
```

```python
# Get all HTML in document and assign it to BeautifulSoup variable
innerHTML = browser.execute_script('return document.body.innerHTML')
soup = BeautifulSoup(innerHTML,"lxml")

# Finds all of the individual products
items = soup.find_all('div', 'tile-container')

# Get all product information on this page
for index in range(len(items)):
    item = items[index]
    item_img = item.find('img')
    item_info = item.find('div','product-info')
    src = item_img.get('src', item_img.get('dfr-src'))
    if src is None or item_info is None:
        continue
    else:
        #item_name = (item.find('div','product-info').find('a',{'target' : '_parent'}).text)
        #item_name = (item.find('div','product-info').find('a').text)
        item_name = item_info.find('a').text        
        data += [{'product-image': src,
                  'product-name' : item_name.replace('\n', '').replace('\t','').strip()}]
```

```python
for i in range(len(data)):
    print(i+1)
    print(data[i]['product-image'])
    print(data[i]['product-name'])
    
browser.quit()
```
