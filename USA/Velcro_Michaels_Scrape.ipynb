```python
from bs4 import BeautifulSoup
from selenium import webdriver

browser = webdriver.Chrome()
url = 'http://www.michaels.com/search?q=velcro#q=velcro&pmpt=qualifying&sz=53'
browser.get(url)
innerHTML = browser.execute_script('return document.body.innerHTML')
soup = BeautifulSoup(innerHTML)

data = []
```

```python
items = soup.find_all("li", class_='grid-tile')

for item in range(len(items)):

    data += [{'product-image': items[item].find(class_='product-image').find('img')['data-original'],
              'product-name' : items[item].find(class_='product-name').find('h3').a.get_text()}]
    
    data[-1]['product-name'] = data[-1]['product-name'].replace('\n', '').replace('\t','')
```
