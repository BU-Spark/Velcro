```python
from bs4 import BeautifulSoup
from selenium import webdriver

# Going to Velcro search page in Amazon
options = webdriver.ChromeOptions()
options.add_argument("--headless")
browser = webdriver.Chrome(chrome_options=options)
#url = 'https://www.amazon.com.au/'
url = 'https://www.amazon.com.au/s/ref=nb_sb_noss_1/356-5976402-9509041?url=search-alias%3Daps&field-keywords=velcro'
browser.get(url)

#browser.find_element_by_id('headerSearch').send_keys('velcro')
#browser.find_element_by_id('headerSearchButton').click()

data = []
```

```python
hasNext = True

while hasNext:

    # Get all HTML in document and assign it to BeautifulSoup variable
    innerHTML = browser.execute_script('return document.body.innerHTML')
    soup = BeautifulSoup(innerHTML,"lxml")
    
    # Finds all of the individual products
    items = soup.find_all('div', 'a-fixed-left-grid-inner')
    
    # Get all product information on this page
    for index in range(len(items)):
        item = items[index]
        item_img = item.find('img','s-access-image')['src']
        item_name_container = item.find('div','a-col-right')        
        item_name = item_name_container.find('h2','s-access-title').text
        data += [{'product-image': item_img,
                  'product-name' : item_name.replace('\n', '').replace('\t','')}]

    # If there is a next page go to the next page
    try:
        nextLink = soup.find("a", {"title" : "Next Page"})
        browser.get('https://www.amazon.com/' + nextLink['href'])
    except:
        hasNext = False
```

```python
for i in range(len(data)):
    print(i+1)
    print(data[i]['product-image'])
    print(data[i]['product-name'])
    
browser.quit()
```
