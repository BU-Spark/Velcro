```python
from bs4 import BeautifulSoup
from selenium import webdriver

browser = webdriver.Chrome()
url = 'https://uk.rs-online.com/web/c/?limit=100&pn=1&r=t&searchTerm=velcro&sra=oss'
browser.get(url)
innerHTML = browser.execute_script('return document.body.innerHTML')
soup = BeautifulSoup(innerHTML, "html5lib")
# browser.close()

data = []
```

```python
items = soup.find_all("div", "tile")

for item in range(len(items)):
    data += [{'product-image' : items[item].find('img')['src'],
              'product-name' : items[item].find('img')['alt'][2:]
             }]
    data[-1]['product-name'] = data[-1]['product-name'].replace('\t','')
```
