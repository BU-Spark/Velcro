```python
from bs4 import BeautifulSoup
from selenium import webdriver

browser = webdriver.Chrome()

page = 1
data = []

while True:
    url = 'https://www.mpdhookandloop.com/search?s=1&keyword=velcro&page=' + str(page)
    browser.get(url)
    innerHTML = browser.execute_script('return document.body.innerHTML')
    soup = BeautifulSoup(innerHTML, "html5lib")
    # browser.close()

    check = soup.find_all('div', 'proditem')
    if len(check) == 0:
        break
    page += 1
    
    items = soup.find_all("div", "proditem")
    for item in range(len(items)):
        data += [{'product-image' : items[item].find('img')['src'],
                  'product-name' : items[item].find('a', 'name').text
                 }]
        data[-1]['product-name'] = data[-1]['product-name'].replace('\t','')
        
```
